```python
def jointpdf(y1, y2, mu1, mu2, sigma1, sigma2):

  fj = 1/(np.sqrt(2 *np.pi) * sigma1 * sigma2) * np.exp(-0.5 * (((y1-mu1)**2 / sigma1**2)+((y2-mu2)**2 / sigma2**2)))


  return fj
```

```python
def marginal(y, mu, sigma):

  f = 1/(np.sqrt(2 *np.pi) * sigma) * np.exp(-0.5 * (y-mu)**2 / sigma**2)

  return f
```

```python
def marginal_cdf(y, mu, sigma):

  f = (1.0)/(2.0) * ( sp.special.erf((y-mu) / (np.sqrt(2)*sigma)) + 1)

  return f
```

```python
def mgf(t, mu, sigma):

  f = np.exp(1.0/2.0 * sigma**2 * t**2 + mu*t)

  return f
```
